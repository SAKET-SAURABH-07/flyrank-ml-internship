## 2. The question: decision, action, cost of a wrong call

*What decision does your work improve? Who acts on it? What does a wrong recommendation cost?*

```python
# #
# #Decision: out of thousands of content pages, which ones should be reviewed first for a
# refresh, expansion, or protection pass?

# Who acts, and what they do: a content editor or SEO lead pulls the top of a ranked review
# queue (weekly, say) and decides — for each page — whether to refresh, expand, leave alone,
# or flag for deeper investigation.

# Cost of a wrong call:
# - False positive (flagged for review, didn't need it): wasted editor hours on a page that
#   was fine.
# - False negative (should have been flagged, wasn't): a page that is quietly losing
#   visibility keeps declining unnoticed, and the traffic/lead loss compounds before anyone
#   catches it.

# Because a missed decline can be more expensive over time than a wasted review hour, the
# ranking should lean toward not missing real decay risk, even at the cost of some false
# positives — but that is a threshold/policy choice to revisit once the model exists, not a
# claim I can make yet.

# Why data or ML helps at all: with a handful of pages, a human could eyeball all of them.
# At thousands of pages across dozens of clients, no editor can manually re-check everything
# every week. A single hand-written rule (e.g. "flag anything not updated in 90 days") is
# also too blunt — it ignores position, demand, and trend direction all at once. A ranked
# score that combines several signals is what makes triage possible at this scale.
# #
# #
```

## 3. Quick look at the data (2-3 real numbers)

*Load the starter CSV below and show 2-3 real numbers that make your lane look worth the next 7 weeks.*

```python
import pandas as pd

df = pd.read_csv("data/raw/content_refresh_anonymized.csv")

visible = df[df["avg_position"] > 0]
print(f"Visible pages (have ranking data): {len(visible):,} of {len(df):,} total rows")

good_positions = ["top_3", "page_1", "striking"]
decay_risk = visible[
    (visible["trend_direction"] == "down") &
    (visible["position_tier"].isin(good_positions))
]
print(f"Good-position pages trending down: {len(decay_risk):,} "
      f"({len(decay_risk) / len(visible) * 100:.1f}% of visible pages)")

impr_at_risk = decay_risk["impressions_90d"].sum()
impr_total = visible["impressions_90d"].sum()
print(f"Impressions sitting in decay-risk pages: {impr_at_risk:,.0f} "
      f"({impr_at_risk / impr_total * 100:.1f}% of all visible-page impressions)")
```
