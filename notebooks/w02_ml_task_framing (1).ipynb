## 3. Success metric

*One metric you can defend. What number means 'good'?*

**Success Metric: Mean Absolute Error (MAE)**

For predicting a continuous variable like `clicks_90d`, **Mean Absolute Error (MAE)** is a highly relevant and interpretable success metric. MAE measures the average magnitude of the errors in a set of predictions, without considering their direction. It is the average over the test sample of the absolute differences between prediction and actual observation where all individual differences have equal weight.

**Why MAE is suitable here:**

1.  **Interpretability:** MAE is in the same units as the target variable (`clicks_90d`), making it easy to understand. An MAE of, for example, 50 means that, on average, our model's predictions are off by 50 clicks.
2.  **Robustness to outliers:** Unlike Root Mean Squared Error (RMSE), MAE does not heavily penalize large errors, making it more robust to outliers in the `clicks_90d` distribution. This is important if some content pieces have exceptionally high or low click counts that might skew other metrics.
3.  **Direct Relevance to Business Goal:** A lower MAE indicates that the model is more accurately predicting the expected clicks, which directly translates to better decision-making for content refresh. If we can reliably predict `clicks_90d`, we can more effectively identify underperforming content that needs a refresh, or high-potential content that could be promoted.

**What number means 'good'?**

A "good" MAE would be a value that is small relative to the typical range and magnitude of `clicks_90d`. The specific target value for MAE would need to be determined in collaboration with domain experts and considering the business impact of prediction errors. For example, if `clicks_90d` typically ranges from 100 to 1000, an MAE of 50 might be considered acceptable, while an MAE of 500 would not. The goal is to minimize MAE as much as possible while balancing model complexity and generalization.

## 5. Why ML beats a fixed rule here

*What makes the pattern too messy for an if-statement?*

Machine Learning offers significant advantages over a fixed rule-based system for predicting content performance (e.g., `clicks_90d`) for several key reasons:

1.  **Complexity and Interdependencies of Factors:** Content performance is not determined by a single factor but by a complex interplay of many variables such as `search_volume`, `competition_level`, `content_type`, `word_count`, `age_tier`, `trend_pct`, and `client_id` specific characteristics. A fixed rule system would struggle to capture the non-linear relationships and intricate interactions among these features. For example, a high `search_volume` content with high `competition` might perform differently based on its `content_type` and `age_tier`. Manually defining rules for every combination and interaction would be intractable and prone to errors.

2.  **Dynamic Nature of Content Performance:** The success of content is not static; it evolves over time due to changing trends, seasonality, competitor activity, and user behavior. Metrics like `trend_direction` and `trend_pct` explicitly highlight this dynamic nature. Fixed rules would require constant, manual updates to adapt to these shifts, leading to high maintenance overhead and potential obsolescence. ML models, particularly those trained on time-series data or with time-sensitive features, can learn and adapt to these evolving patterns automatically.

3.  **High Dimensionality and Scale:** The dataset contains numerous features (44 columns), many of which are continuous or categorical with multiple levels. Creating a comprehensive rule-based system that considers all these dimensions and their potential thresholds would be overwhelmingly complex. As the content catalog grows, the complexity of managing and scaling such a rule set would quickly become unmanageable.

4.  **Identifying Latent Patterns:** ML algorithms can uncover subtle, non-obvious patterns and relationships within the data that are not immediately apparent to human experts or easily expressible through simple `if-else` statements. These latent patterns might provide crucial insights into content performance drivers that fixed rules would invariably miss.

5.  **Optimization for a Specific Metric:** ML models can be explicitly trained and optimized to minimize the Mean Absolute Error (MAE) of `clicks_90d` predictions. Fixed rules, by contrast, are typically designed to satisfy specific conditions, making it difficult to optimize directly for a continuous numerical target and achieve the best possible predictive accuracy.
